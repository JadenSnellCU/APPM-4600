```python
#Driver Code + Exercise 1
import numpy as np
import numpy.linalg as la
import math
def driver():
    n = 100
    x = np.linspace(0,2*np.pi,n)
    
    f = lambda x: np.cos(x)
    g = lambda x: np.sin(x)
    
    y = f(x)
    w = g(x)  
    
    dp = dotProduct(y,w,n)
    print('the dot product is : ', dp)
    return
def dotProduct(x,y,n):
    dp = 0.
    for j in range(n):
        dp = dp + x[j]*y[j]
    return round(dp,10)
driver()
```

```python
#Making a matrix/Exercise 
def mat_mult(mat,vec):
    num_rows = len(mat)
    product = np.array(np.zeros(num_rows))
    for i in range(num_rows):
        product[i] = dotProduct(mat[i],vec,len(mat))
            
    return product
mat = np.array([[1, 2 ,3, 4],[6,7 ,8,9],[1, 2 ,3, 4],[6,7 ,8,9]])
vec = np.array([10, 11,12,13])
product = mat_mult(mat,vec)
print(product)
            
```
